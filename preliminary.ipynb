```python
from openai import OpenAI

# Define a function to generate prompts based on demographic profiles
def generate_prompt(profile):
    sex, age_range, activity = profile
    return f"Generate a daily schedule for a {age_range}-year-old {sex} who is a {activity}. Include activities such as sleeping, working, cooking, and watching TV. Specify the start time, end time, and duration for each activity."

# Define demographic profiles (sex, age range, professional activity)
profiles = [
    ("male", "30-39", "working person"),
    ("female", "20-29", "student"),
    # Add more profiles based on processed TUS for SMACH
]

# Generate prompts for each profile
prompts = [generate_prompt(profile) for profile in profiles]

# Initialize the OpenAI client to point to the local LM Studio server
client = OpenAI(base_url="http://localhost:1234/v1", api_key="lm-studio")

# Function to get schedule using LM Studio's OpenAI-like interface
def get_schedule(prompt):
    completion = client.chat.completions.create(
        model="lmstudio-community/Meta-Llama-3-8B-Instruct-GGUF",
        messages=[
            {"role": "system", "content": "Generate daily activity schedules."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.7,
    )
    # Print the structure of the response
    print(completion)

    # Extracting the content correctly
    return completion.choices[0].message.content

# Generate schedules for each prompt
schedules = [get_schedule(prompt) for prompt in prompts]

# Print the generated schedules
for profile, schedule in zip(profiles, schedules):
    print(f"Profile: {profile}")
    print(f"Schedule: {schedule}")
    print("\n")

# Test the code
if __name__ == "__main__":
    for i, (profile, schedule) in enumerate(zip(profiles, schedules)):
        print(f"Profile {i+1}: {profile}")
        if schedule:
            print(f"Generated Schedule:\n{schedule}\n")
        else:
            print("No schedule generated.\n")
```
